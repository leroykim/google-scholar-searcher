```python
import pandas as pd
```

```python
veracity_df = pd.read_excel('results/data veracity.xlsx', index_col=0)
credibility_df = pd.read_excel('results/data credibility.xlsx', index_col=0)
trustworthiness_df = pd.read_excel('results/data trustworthiness.xlsx', index_col=0)
```

```python
credibility_df['pub_year'] = credibility_df['pub_year'].fillna(0)
credibility_df['pub_year'] = credibility_df['pub_year'].astype(int)
```

```python
veracity_df['eprint_url'] = veracity_df['eprint_url'].fillna("")
credibility_df['eprint_url'] = credibility_df['eprint_url'].fillna("")
trustworthiness_df['eprint_url'] = trustworthiness_df['eprint_url'].fillna("")
```

```python
all_df = pd.concat([veracity_df, credibility_df, trustworthiness_df], ignore_index=True)
```

```python
all_df = all_df.drop_duplicates(subset=['title'], ignore_index=True)
```

```python
all_df = all_df.sort_values(by=['num_citations'], ascending=False, ignore_index=True)
```
